```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[45:55]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.30)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'rbf', C = 30, gamma = 'auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# K_Fold to check the score of SVC with different parameters like kernel, C and gamma

from sklearn.model_selection import cross_val_score
cross_val_score(SVC(kernel = 'rbf', C = 30, gamma = 'auto'), iris.data, iris.target)
```

```python
cross_val_score(SVC(kernel = 'linear', C = 30, gamma = 'auto'), iris.data, iris.target)
```

```python
cross_val_score(SVC(kernel = 'rbf', C = 20, gamma = 'auto'), iris.data, iris.target)
```

```python
import numpy as np
kernel = ['rbf', 'linear']
C = [1, 10, 20]
avg_score = {}
for kval in kernel:
    for cval in C:
        cv_scores = cross_val_score(SVC(kernel=kval, C=cval, gamma='auto'), iris.data, iris.target, cv = 5)
        avg_score[kval + '_' + str(cval)] = np.average(cv_scores)

avg_score
```

```python
# the above for loop thing with GridSearchCV

from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(SVC(gamma = 'auto'), {
    'C' : [1, 10, 20, 30],
    'kernel' : ['rbf', 'linear']
}, cv=5, return_train_score=False)

clf.fit(iris.data, iris.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
# The GridSearchCV can require more computation power for large datasets
# So to tackle this we have RandomizedSearchCV that randomly takes the parameter and run the model for specified iterations

from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(SVC(gamma = 'auto'), {
    'C' : [1, 10, 20, 30],
    'kernel' : ['rbf', 'linear']
},  cv = 5,
    return_train_score=False,
    n_iter=3)

rs.fit(iris.data, iris.target)
pd.DataFrame(rs.cv_results_)[['param_C', 'param_kernel', 'mean_test_score']]
```
